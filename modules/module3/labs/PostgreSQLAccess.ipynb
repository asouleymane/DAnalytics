## Basic Queries in SQL

First we will discuss the basic Select statement again. 

        SELECT * FROM cities;
        
This statement will return all of the data from the given table (cities). This is true in all SQL select statments that * means return all the attributes of a table. 

Now let's limit what we return from the table. This is easily done by replacing the * in the select statment with the attributes you want to return seperated by commas.

Run the following select statement and see the different between it and the select * statement we ran earlier.

        SELECT city, country FROM cities;
        
Now let's discuss the where clause again. This clause allows us to limit what is returned in the select statement by adding a WHERE on to the end.

The following statement will return only the cities name of cities that are located in the United States.

        SELECT city FROM cities WHERE country = 'United States';
        
Where clauses are useful when you are looking for certain portions of the data that is located in the table. 

But what if we wanted all the cities in the United States OR Canada. 

We can impliment a OR clause which will allow our where clause to recognize both countries

        SELECT city FROM cities WHERE country = 'United States' OR country = 'Canada';
        
We can also use an AND clause to see cities in the United States AND has a population that is greater than 2500000.

        SELECT city FROM cities WHERE country = 'United States' AND population > 2500000;

You can easily order results from the query by adding a ORDER BY condition. 

This will allow you to choose an attribute of the table and order the data based on that atribute in ascending or descending order.

        SELECT city FROM cities WHERE country = 'United States' AND population > 2500000 ORDER BY city ASC;
        
This query will return the same results as before, but this time the results will be organized by the city name in ascending order. 

To change it to descending order simply replace ASC with DESC in the query.

Take some time here to query the data adding simple constraints and learning how the SQL languague works.
        

## Updating / Read Only Status

As we learned in the last activity you can update data in the table by using the UPDATE command. Enter the following command:

        UPDATE us_second_division SET state_number_code = 2 WHERE state_number_code = 1;
        
This should take all the counties from the second state and add them to the first, but you will receive a error statement that says permission denied.

This is because you are logged into a read only account which allows you to query the data, but not insert, delete or update the tables. 

You may wonder why accounts like this are neccessary. It allows different users to query the database, but not be given the right to change the data for risk of someone messing up the data.
